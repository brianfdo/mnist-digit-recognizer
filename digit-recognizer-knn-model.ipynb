```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import csv
```

```python
train_data = pd.read_csv("../input/digit-recognizer/train.csv")
test_data = pd.read_csv("../input/digit-recognizer/test.csv")

print(train_data.shape)
print(test_data.shape)

train_data.head()
```

```python
labels = train_data["label"]
train_values = train_data.drop(labels = ["label"], axis = 1)

print(train_values.shape)
print(train_values.head())
```

```python
train_values.isnull().any().describe()
```

```python
test_data.isnull().any().describe()
```

```python
train_images,test_images,train_labels,test_labels = train_test_split(train_values,labels,random_state = 2,train_size = 0.8)
```

```python
knn = KNeighborsClassifier(n_neighbors=3, weights='distance', algorithm='auto', leaf_size=30, p=2, metric='minkowski')
knn.fit(train_images,train_labels.values.ravel())
```

```python
predictions = knn.predict(test_images)

from sklearn.metrics import accuracy_score
print(accuracy_score(test_labels,predictions))
```

```python
test_pred = knn.predict(test_data)

data_frame = pd.DataFrame(test_pred)
print(data_frame.head())
```

```python
data_frame.index.name = 'ImageId'
data_frame.index += 1
data_frame.columns = ['Label']
data_frame.to_csv('digitrecognizer.csv', header = True)

print(data_frame.head())
print(data_frame.size)
```
